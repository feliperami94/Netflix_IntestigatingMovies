```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
#print(netflix_df)
movies_rel_1990s = netflix_df[(netflix_df['release_year']>=1990) & (netflix_df['release_year']<=1999)] #These are a subset of movies released in the 1990s
#print(movies_rel_1990s)
```

```python
#Plotting a histogram of the movies duration to see the most frequent duration
plt.hist(movies_rel_1990s['duration'])
plt.show()
duration = 100
```

```python
#Counting the number of short action movies from the 1990s
#First subsetting the action ones
action_mov_1990s = movies_rel_1990s[movies_rel_1990s['genre'] == 'Action']
#Setting a counter to find the number of movies with less than 90 minutes
short_movie_count = 0

for label, row in action_mov_1990s.iterrows():
    if row['duration'] < 90:
        short_movie_count = short_movie_count + 1

#print(short_movie_count)
```
